## Separation of returns: 超额收益率的勾股定理

对一个投资产品来说，它的超额收益率$$r_p = \beta r_M + \alpha,$$ 其中 $\alpha \perp r_M$:
- $Beta$: 和总体市场的超额收益率平行的部分 $\beta r_M$
- $Alpha$: 和总体市场的超额收益率垂直的部分 $\alpha$, called idiosyncratic returns
