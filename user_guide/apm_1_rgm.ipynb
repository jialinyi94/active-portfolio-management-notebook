# Active Portfolio Management 1: Return Generative Model

## Excess Returns = Alpha + Beta + Noise

:::{important}
Alpha: 某一个资产所独有的, 其他的资产所没有的特征
:::

:::{important}
Beta: 同时影响所有资产的收益率的特征
:::

两者的区别在于: Alpha是资产独有的特征, 而Beta是所有资产共有的特征. 例如, AAPL的Alpha可能是公司的CEO是Tim Cook, 而Beta可能是整个市场的波动率. 因此, Alpha和Beta的区别在于它们的影响范围和特性.

:::{important}
Noise: 随机的, 不可预测的收益率
:::

这三者在维度上也有所不同:
- Alpha: 1维，维度是不同的asset
- Beta: 2维，维度是
