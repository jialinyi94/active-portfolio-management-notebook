# Technical Appendix: Mean-Variance analysis

下面我按 “一句直觉 + 一段数学骨架 + 你在做投资时怎么用”的方式，把第二章后面的 Technical Appendix（2A）讲清楚。

核心就一件事：把“一个特征/观点（attribute）”和“一个最有效表达它的组合（portfolio）”一一对应起来，这套 machinery 后面会反复用到（alpha、beta、Q、A、信息比率…都从这里长出来）

## 1) 特征组合：满足条件的最小方差投资组合

我们可以用一个向量来表示所有资产的某一个特征
$$
a = (a_1,\dots,a_N)^\top
$$

然后问一个非常 Markowitz 的问题：

:::{note}
如果我们想构建资产组合，配置出“1单位该特征”且方差最小，该怎么建仓？
:::

比如：
- $a=e=(1,1,\dots,1)$：如果我想把钱全用来投资
- $a=f$ 表示“预期超额收益”：如果我希望预期超额收益为1
- $a=\beta$ 表示“相对某个基准的 beta”: 如果我希望控制资产组合的beta=1
- $a=$ E/P、size、行业哑变量……都可以

这就是 Appendix 里定义的 Characteristic Portfolio（特征组合）。 

## 2) Proposition 1：特征组合的通用解

- 2.1 定义问题（一个带线性约束的二次规划）
$$
\min_h\; h^\top V h \quad \text{s.t.}\quad h^\top a = 1
$$
其中$V$是资产收益的协方差矩阵。 ￼

- 2.2 一阶条件直接给出 closed-form

Lagrangian：$$L(h,\delta)=h^\top V h-\delta(h^\top a-1)$$

FOC 解出来就是书上式 (2A.1)/(2A.9) 的结论：
$$
h_a \;=\;\frac{V^{-1}a}{a^\top V^{-1}a}
$$

:::{important}
你想“押注/表达”某个方向 a，最省风险的做法就是用 $V^{-1}$ 把相关性和波动率“去噪/去耦合”。 ￼
:::

- 2.3 这个组合的方差也很漂亮
$$
\sigma_a^2 = h_a^\top V h_a = \frac{1}{a^\top V^{-1}a}
$$
这就是 (2A.2)。 ￼

## 3) 最关键的魔法：“暴露”=“beta”

Appendix 的一个很强的“翻译器”是式 (2A.3)：

资产$n$对特征组合 $h_a$ 的 beta，等于该资产的特征值 $a_n$。

形式上就是
$$
\beta_{n|h_a}=\frac{\mathrm{Cov}(r_n, r_{h_a})}{\mathrm{Var}(r_{h_a})}=a_n
$$
这就是 (2A.3) 的意思。 ￼

你把“特征”做成了一个组合后，市场里任何资产“在这个特征上的含量”，就能用一个完全金融化的量来读——beta。

所以从此以后，“因子暴露/特征暴露”这件事，可以完全在组合语言里玩。


## 4) 三个重要例子：C、B、Q（以及 q）

### 4.1 Portfolio C：最小方差的“满仓组合”

取特征 $a=e=(1,\dots,1)$。约束 $h^\top e=1$ 就是“权重加总为 1（全投资）”。

因此 $C$ 是“所有全投资组合里方差最小”的那个：式 (2A.14)。 ￼

并且有一个漂亮结论：每个资产相对于 C 的 beta 都等于 1（式 (2A.16)）。书里还给了直觉：$C$ 已经最小方差了，所以每个资产对组合风险的边际贡献必须一致，否则还能再 swap 降风险；一致 + 自身 beta=1 推出每个资产 beta=1。 ￼

### 4.2 Portfolio B：基准就是“beta 这个特征”的特征组合

如果你把“beta”这个特征定义为“相对于某个基准组合 B 的 beta”，那么 Appendix 证明：基准组合本身，就是 beta 的特征组合（式 (2A.19) 一带）。 ￼

一句话：你拿什么定义 beta，那个东西本身就被“内生地”放回到这套特征组合体系里了。

### 4.3 Portfolio $q$ & $Q$：预期收益向量 $f$ 的特征组合，以及“最大 Sharpe”

把特征取成 $a=f$（预期超额收益向量），得到它的特征组合 $q$。书里接着把它和 最大 Sharpe 连起来：
因为 Sharpe 比例对缩放不敏感，所以“最大 Sharpe”可以等价写成“固定期望收益=1，最小方差”，这正是 $f$ 的特征组合问题。 ￼

然后在一些条件下（尤其是 $f_C$>0），可以把 $q$ 归一化成一个全投资的版本 Q（式 (2A.33)-(2A.36)）。最重要的就是 (2A.36)：
$$
f = f_Q \cdot \beta(\cdot\,\text{w.r.t. }Q)
$$
也就是：每个资产的预期超额收益，和它对组合 Q 的 beta 成正比。 ￼

:::{important}
Mean-variance analysis给了一个非常几何的结论：
最大Sharpe的组合是天然的定价因子，每个投资组合都可以用它在最大Sharpe组合上的投影进行定价
:::

:::{note}
CAPM因此等价于:
- 每个投资者面对同一套$f$和$V$
- 市场组合就是最大Sharpe组合
:::

这就是这章要把 CAPM 讲成“共识预期收益”的核心数学骨架：CAPM 并不是先验神谕，而是说——存在一个组合 $Q$，使得共识的 $f$ 看起来像“单因子 beta 定价”。

## 5) 你在实操里怎么用这套 machinery（非常值钱）

把它当成一个“从信息/约束到组合构造的模块化积木”：

1.我有一个我要表达的东西（特征向量 $a$：某个 alpha、某个因子暴露、某个行业倾向、某种天气风险暴露…）

2.我希望表达它的同时风险最省
直接用 $$h_a = \frac{V^{-1}a}{a^\top V^{-1}a}$$（这就是“最信息效率/最风险效率”的原型）

3.我希望再加上“全投资”“beta=1”“不碰某些风险”之类的约束
后面章节会把这变成在特征组合空间里做线性组合（Appendix 末尾也点到：A、B、
C 线性组合来满足多种属性） ￼

如果你把它放进你熟悉的量化语言：
- $V^{-1}$ 是“risk-adjusted 的几何”
- $a$ 是“你想下注的方向”
- $h_a$ 是“在相关性世界里，最便宜的下注方式”

## 6) Efficient frontier

如果我希望构建一个全投资的资产组合，满足预期超额收益=$\lambda$，且方差最小。通过求解下面这个最优化问题，可以得到
$$
\min_h\; h^\top V h \quad \text{s.t.}\quad h^\top a = 1, h^\top f = \lambda
$$

$$
h = \frac{f_Q - \lambda}{f_Q - f_C} h_C + \frac{\lambda - f_C}{f_Q - f_C} h_Q
$$
其中
- $Q$是当$\lambda=1$的解，也就是Sharpe ratio最大的组合
- $C$是全投资的方差最小的投资组合
- $f$和$h$分别代表预期超额收益率和组合头寸
