# APM 101 - 1: Return Generative Model

## Excess Returns = Alpha + Beta + Noise

:::{important}
Alpha: 某一个资产所独有的, 其他的资产所没有的特征
:::

:::{important}
Beta: 同时影响所有资产的收益率的特征
:::

两者的区别在于: Alpha是资产独有的特征, 而Beta是所有资产共有的特征. 例如, AAPL的Alpha可能是公司的CEO是Tim Cook, 而Beta可能是整个市场的波动率. 因此, Alpha和Beta的区别在于它们的影响范围和特性.

## A statistical generative model for returns

$$
x_{i,t} = \alpha_i + \beta_{i, k} f_{k,t} + \epsilon_{i,t}
$$

- $x_{i,t}$: 资产i在时间t的收益率
- $\alpha_i$: 资产i的Alpha
- $\beta_{i, k}$: 资产i对因子k的敏感度
- $f_{k,t}$: 因子k在时间t的值

这个模型等价于一个深度学习中的Linear layer, 其中batch_dim是time, in_feature是factors, out_features是assets
