# Chapter 2: Consensus Expected Returns: CAPM

## Separation of returns: 超额收益率的勾股定理

对一个投资产品来说，它的超额收益率$$r_p = \beta_p r_M + \alpha,$$ 其中 $\alpha \perp r_M$:
- $Beta$: 和总体市场的超额收益率平行的部分 $\beta_p r_M$
- $Alpha$: 和总体市场的超额收益率垂直的部分 $\alpha_p$, called idiosyncratic returns

那么，勾股定理告诉我们，投资产品的风险可以分解成为两个不相关的风险源:
$$
\sigma_p^2 = \beta_p^2 \sigma_M^2 + \omega_p^2
$$
其中:
- $\sigma_M$ 是总体市场的风险
- $\omega_p$ 是投资产品所特有的风险

## Captial Asset Pricing Model: 所有的Alpha只是运气

CAPM的拥护者认为: 对任何的投资产品$p$来说，$$\mathbb{E}[\alpha_p] = 0$$

:::{note}
CAPM认为，不存在Alpha空间。不需要主动投资，被动投资就是最好的投资方式。
:::

如果CAPM成立，那么所有人都会有一样的投资预期$\beta_p r_M$, 那么每个人只需要持有市场组合和无风险债券即可，具体的比例根据自己的风险偏好来调，这就是passive portfolio management

:::{important}
被动投资对应的收益率，$\beta_p r_M$ 就是 consensus expected returns.
:::

这和有效市场假说是一致的，Efficient Market Hypothesis的三种不同的程度的假说
- weak: 只依赖price和volume信息的投资者无法打败市场组合
- semi-weak: 只依赖公开信息的投资者无法打败市场组合
- strong: 任何投资者都无法击败市场组合

分别对应：
- weak: price和volume信息里不存在alpha
- semi-weak: 公开信息中不存在alpha
- strong: 不存在alpha, 或者无法预测 <- CAPM的支持者在这里

## Difference between active and passive portfolio management: 对alpha的预测能力

Passive portfolio management
- 投资经理都对Alpha所在的空间一无所知，也无从预测
- $\mathbb{E}[\alpha_p] = 0$
- 只需根据风险偏好调整市场组合和无风险债券的比例
- 任务是预测beta

Active portfolio management
- 投资经理具有预测Alpha的能力
- $\mathbb{E}[\alpha_p] \neq 0$
- 最优投资组合不再是市场组合和无风险债券的线性组合
- 任务是控制beta，预测alpha.
